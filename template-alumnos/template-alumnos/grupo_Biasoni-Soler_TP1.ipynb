### Pregunta 2 -  ¬øC√≥mo se garantiza existencia de la factorizaci√≥n $LU$? ¬ø La matriz $(\mathbf{I} - p \ \mathbf{W} \ \mathbf{D})$ est√° bien condicionada? ¬øC√≥mo influye el valor de $p$?

### ¬øC√≥mo se garantiza existencia de la factorizaci√≥n  ùêøùëà?  


Se dice que una matriz cuadrada es *diagonal dominante* (por filas) si el valor absoluto de la entrada en la diagonal principal de una fila es mayor o igual a la suma de los valores absolutos de todas las dem√°s entradas (no diagonales) de esa fila.

Para probar que existe la factorizacion LU de la matriz A= I-pWD, comenzaremos probando que su trasnpuesta es estrictamente diagonal dominante.
Para ello queremos ver que :

<div style="margin-bottom:20px;text-align:left">
$\forall j \in [1,n]   \sum_{i=1}^{n} |-p(WD)_{ij}| <1$
</div>
<div style="margin-bottom:20px;text-align:left">
$\sum_{i=1}^{n} |-p(WD)_{ij}| = p\sum_{i=1}^{n} (WD)_{ij}$ 
</div>
    
<div style="margin-bottom:20px;text-align:left">
\((WD)_{ij} = 
\begin{cases}
    0 & \text{si } c_{j} = 0 \\
    \dfrac{W_{ij}}{c_j} & \text{si } c_{j} \neq 0
\end{cases} \)
</div>

<div style="margin-bottom:20px;text-align:left">
$\text{si } c_{j} = 0$
</div>
<div style="margin-bottom:20px;text-align:left">    
$p\sum_{i=1}^{n} 0 <1$ 
</div>
<div style="margin-bottom:20px;text-align:left">    
$\text{si } c_{j} \neq 0$
</div>
<div style="margin-bottom:20px;text-align:left">    
$\dfrac{p}{c_j}$ $\sum_{i=1}^{n} W_{ij}$
</div>
<div style="margin-bottom:20px;text-align:left">    
Sabemos que $c_j$ = $\sum_{i=1}^{n} W_{ij}$
</div>
<div style="margin-bottom:20px;text-align:left">    
$\rightarrow \dfrac{p}{c_j}c_j = p$
</div>
<div style="margin-bottom:20px;text-align:left">    
$\text{si } c_{j} \neq 0$
</div>
<div style="margin-bottom:20px;text-align:left">    
$\sum_{i=1}^{n} |-p(WD)_{ij}| = p$
</div>
<div style="margin-bottom:20px;text-align:left">    
Como $p \in (0,1)$
</div>    
$p <1 \forall p$ 
</div>
$\rightarrow(I-pWD)^t$ es EDD
    
Como probamos que $(I-pWD)^t$ es EDD, sabemos que tiene descomposicion LU y es inversible (pues U inversible)

Luego, $det[(I-pWD)^t)] \neq 0$ y $det[(I-pWD)^t)] = det[(I-pWD)] $ 

$\rightarrow$ I-pWD es inversible

Como $(I-pWD)^t$ es inversible, y tiene LU $\rightarrow$ sus menores principales son $\neq$ 0. Por lo tanto, los menores principales de I-pWD tambien lo son (son iguales)

Como vimos que I-pWD es inversible y tiene menores principales $\neq$ 0 $\rightarrow$ tiene factorizacion LU

Queremos probar que teniendo una matriz $A = I - pWD$ , √©sta admite descomposicion LU, si A es diagonal dominante.
